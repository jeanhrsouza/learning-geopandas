```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
```

```python
censo = pd.read_excel('../01.Dados/Censo 2010/RJ/EXCEL/Basico_RJ.xls')
```

```python
setor = gpd.read_file('../01.Dados/Mapas/RJ-SETOR/RJ-SETOR.shp')
```

```python
censo.head()
```

```python
setor.head()
```

### Igualando o tipo de variável 

```python
# Identificando o tipo de variável que deseja trabalhar (Cod_Setor)
censo.dtypes
```

```python
# Identificando o tipo da variável (CD_GEOCODI)
setor.dtypes
```

```python
# Trocando o tipo de variável 
censo['Cod_setor'] = censo['Cod_setor'].astype(str)
```

```python
censo.dtypes
```

```python
print(type(censo))
print(type(setor))
```

```python
setor.shape
```

```python
# Realizando a junção [merge] de DataFrames;
# Reparar que existe uma variável em comum. [lembrar de chave estrangeira em SQL]. No caso do pandas é o 
# right_on e left_on
# how -> Indica que o dataframe da esquerda não pode perder nenhum dado e o da direita junta com o da
# esquerda nos dados condizentes.
setor = pd.merge(setor, censo, left_on = 'CD_GEOCODI', right_on = 'Cod_setor', how = 'left')
```
