# Criando Shapefiles

## Obtendo os dados

ftp://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_municipais/municipio_2017/UFs/RJ/

## Lendo

```python
%matplotlib inline
import geopandas as gpd
```

```python
rj = gpd.read_file('../01.Dados/Mapas/RJ/33MUE250GC_SIR.shp')
```

```python
rj
```

```python
# Shapefile do estado do Rio de Janeiro e todos os seus Municípios
# color -> fundo
# edgecolor -> limites 
rj.plot(color = 'white', edgecolor = 'black', figsize = (15, 8))
```

## Modificando

```python
rj = rj[rj['NM_MUNICIP'] == 'RIO DE JANEIRO']
```

```python
rj
```

```python
rj.plot(color = 'orange', edgecolor = 'black', figsize = (15, 8))
```

## Salvando 

```python
import os

dir = '../01.Dados/Mapas/RJ-MUNIC'
if not os.path.exists(dir):
    os.makedirs(dir)
```

```python
rj.to_file(dir + '/RJ-MUNIC.shp')
```

```python
# testando se o arquivo foi criado corretamente
rj_munic = gpd.read_file(dir + '/RJ-MUNIC.shp')
```

```python
rj_munic.plot(color = 'orange', edgecolor = 'black', figsize = (15, 8))
```
